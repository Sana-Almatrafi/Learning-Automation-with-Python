```python
import pandas as pd
```

**1- Read a csv file from a URL with Pandas**

target website: https://www.football-data.co.uk/data.php

```python
#reading 1 csv file from the website
url = 'https://www.football-data.co.uk/mmz4281/2526/SC0.csv'
df_premier26 = pd.read_csv(url)
```

```python
#printing dataframe
df_premier26
```

```python
#rename columns
df_premier26.rename(columns={'FTHG':'home_goals',
                             'FTAG':'away_goals'}, inplace=True)
```
