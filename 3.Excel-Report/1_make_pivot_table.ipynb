```python
import pandas as pd
```

```python
df = pd.read_excel('/content/supermarket_sales.xlsx')
print(df)
```

```python
df = df[['Gender', 'Product line', 'Total']]
print(df)
```

```python
ptable = df.pivot_table(index='Gender', columns='Product line', values='Total', aggfunc='sum').round(0)
print(ptable)
```

```python
ptable.to_excel('pivot_table.xlsx', 'Report', startrow=4)
```
